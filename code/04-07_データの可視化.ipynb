# データの可視化

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 7, 4, 8])
```

```python
plt.plot(x, y, color='r')
plt.show()
```

## 散布図

```python
plt.scatter(x, y, color='r')
```

## 曲線

```python
x = np.linspace(-5, 5, 300)
y = x ** 2
```

```python
plt.plot(x, y, color='r')
plt.show()
```

## 複数の線を表示

```python
x = np.linspace(-5, 5, 300)
y1 = x ** 2
y2 = (x - 2) ** 2
```

```python
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='k', linestyle='--')
plt.show()
```

## ヒストグラム

```python
np.random.seed(0)
l = []
for _ in range(1000):
    # サイコロを10回投げた合計の値をリストに格納
    l.append(np.random.randint(1, 7, size=10).sum())

plt.hist(l, 20, color='gray')
plt.show()
```

```python
plt.hist(l, np.arange(15, 55, 2))
plt.show()
```

## 複数のグラフを並べて表示

```python
x = np.linspace(-5, 5, 300)
sin_x = np.sin(x)
cos_x = np.cos(x)

# 2行1列にサブプロットを並べる
fig, axes = plt.subplots(2, 1)

# y軸の描画範囲を設定
axes[0].set_ylim([-1.5, 1.5])
axes[1].set_ylim([-1.5, 1.5])

# グラフの描画
axes[0].plot(x, sin_x, color='r')
axes[1].plot(x, cos_x, color='k')
```

## 等高線の描画

```python
def f(x, y):
    return x**2 + y**2 / 4
```

```python
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)

xmesh, ymesh = np.meshgrid(x, y)
z = f(xmesh.ravel(), ymesh.ravel()).reshape(xmesh.shape)

plt.contour(x, y, z, colors='k', levels=[1, 2, 3, 4, 5])
plt.show()
```
